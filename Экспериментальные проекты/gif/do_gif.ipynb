```python
import imageio as io
from IPython.display import Image
from PIL import Image as Im
import os

base_dir = "D:/Projects/Jupyter trash/do gif/"
num_photos = len(os.listdir(base_dir + "photos"))

def do_gif(num_photos, time):  
    images = []
    for i in range(1, num_photos+1, 1):
        images.append(io.imread(base_dir + "resized_photos/" + str(i) + ".jpg"))
    io.mimsave('result.gif', images, duration = time)
    display(Image("result.gif", width=200, unconfined=True))

def resize_im (num_photos, basewidth):
    for i in range(1, num_photos+1, 1):
        img = Im.open(base_dir + "photos/" + str(i) + ".jpg")
        wpercent = (basewidth/float(img.size[0]))
        hsize = int((float(img.size[1])*float(wpercent)))
        img = img.resize((basewidth,hsize), Im.ANTIALIAS)
        img.save(base_dir + "resized_photos/" + str(i) + ".jpg")
    
resize_im(num_photos, 200)
do_gif(num_photos, 0.15)
```

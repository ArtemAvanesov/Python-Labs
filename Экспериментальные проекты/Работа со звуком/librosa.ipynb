# Read musics

```python
import librosa

file_path = 'music.wav'
samples, sampling_rate = librosa.load(file_path, sr=None, mono=None, offset=0.0, duration=None)

print("len samples: {}".format(len(samples)))
print("sampling_rate: {} per second".format(sampling_rate))
```

# Getting duration musics

```python
duration_of_sound = len(samples)/sampling_rate
print("duration: {:.2f} second".format(duration_of_sound))
```

# Playing music

```python
from IPython.display import Audio
Audio(file_path)
```

# Visualizing Audio

```python
import matplotlib.pyplot as plt
from librosa import display

plt.figure(figsize=(10,1))
librosa.display.waveplot(y = samples, sr = sampling_rate)
plt.xlabel("Time, sec")
plt.ylabel("Amplitude")
```

# Simple Sine Wave to Understand FFT

```python
import numpy as np
import warnings
warnings.filterwarnings('ignore')

samples2 = 100
f = 3
x = np.arange(samples2)
y1 = np.sin(2*np.pi*f * (x/samples2))

plt.figure(figsize=(10,2))
plt.stem(x, y1, 'r',)
plt.plot(x, y1)
plt.xlabel("Time")
plt.ylabel("Amplitude")
```

```python
import scipy

def fft_plot(audio, sampling_rate):
    n = len(audio)
    T = 1/sampling_rate
    yf = scipy.fft(audio)
    xf = np.linspace(0.0, 1.0/(2.0*T), int(n/2))
    fig, ax = plt.subplots()
    ax.plot(xf, 2.0/n * np.abs(yf[:n//2]))
    plt.grid()
    plt.xlabel("Frequency")
    plt.ylabel("Magnitude")
    return plt.show
```

```python
fft_plot(y1, samples2)
```

```python
samples3 = 100
f = 11
x = np.arange(samples3)
y2 = np.sin(2*np.pi*f * (x/samples3))

plt.figure(figsize=(10,2))
plt.stem(x, y2, 'r',)
plt.plot(x, y2)
plt.xlabel("Time")
plt.ylabel("Amplitude")
```

```python
fft_plot(y2, samples3)
```

```python
y3 = y1+y2

plt.figure(figsize=(10,2))
plt.stem(x, y3, 'r',)
plt.plot(x, y3)
plt.xlabel("Time")
plt.ylabel("Amplitude")
```

```python
fft_plot(y3, samples3)
```

# FFT of our music

```python
fft_plot(samples, sampling_rate)
```

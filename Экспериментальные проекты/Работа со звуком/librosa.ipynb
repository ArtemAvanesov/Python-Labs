```python
import librosa

file_path = 'music.wav'
samples, sampling_rate = librosa.load(file_path, sr=None, mono=None, offset=0.0, duration=None)

print("len samples: {}".format(len(samples)))
print("sampling_rate: {} per second".format(sampling_rate))
```

```python
duration_of_sound = len(samples)/sampling_rate
print("duration: {:.2f} second".format(duration_of_sound))
```

```python
from IPython.display import Audio
Audio(file_path)
```

```python
import matplotlib.pyplot as plt
from librosa import display

plt.figure(figsize=(10,1))
librosa.display.waveplot(y = samples, sr = sampling_rate)
plt.xlabel("Time, sec")
plt.ylabel("Amplitude")
```

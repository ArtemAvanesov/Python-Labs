```python
import numpy as np

data = np.loadtxt('ionosphere.csv', delimiter=';', unpack=False)
X = data[:, :34]
y = data[:, 34:35]
y = y.reshape(351)

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)

X_train_scaled = X_train
X_test_scaled = X_test

import pandas as pd
from pandas import plotting 
import matplotlib.pyplot as plt
%matplotlib inline

dataframe = pd.DataFrame(X)
dataframe = pd.DataFrame(X_train_scaled)
grr = plotting.scatter_matrix(dataframe, c=y_train, figsize=(50, 50), marker='o', hist_kwds={'bins': 20}, s=30, alpha=.8)
plt.savefig("figure_"+s+".png")
```

```python
#LogisticRegression best
from sklearn.linear_model import LogisticRegression
import pandas as pd

Lr1 = LogisticRegression(C = 0.6, penalty='l1', solver='liblinear', random_state=1)
Lr1.fit(X_train_scaled,y_train)
print("LR_l1.score train: {:.2f}%".format(Lr1.score(X_train_scaled, y_train)*100))
print("LR_l1.score test: {:.2f}%\n".format(Lr1.score(X_test_scaled, y_test)*100))

np_resuts = np.around(np.array(Lr1.coef_), decimals=2)
data_pd = pd.DataFrame(data=np_resuts, index=['coef L1'])
#data_pd.columns = params
display(data_pd)

Lr2 = LogisticRegression(C = 1, penalty='l2', solver='liblinear', random_state=1)
Lr2.fit(X_train_scaled,y_train)
print("LR_l2.score train: {:.2f}%".format(Lr2.score(X_train_scaled, y_train)*100))
print("LR_l2.score test: {:.2f}%\n".format(Lr2.score(X_test_scaled, y_test)*100))

np_resuts = np.around(np.array(Lr2.coef_), decimals=2)
data_pd = pd.DataFrame(data=np_resuts, index=['coef L2'])
#data_pd.columns = params
display(data_pd)
```

```python
# LinearSVC best
from sklearn.svm import LinearSVC
import mglearn


linear_svm_l1 = LinearSVC(C = 1, penalty="l1", dual=False)
linear_svm_l1.fit(X_train_scaled,y_train)
print("linear_svc_l1.score train: {:.2f}%".format(linear_svm_l1.score(X_train_scaled, y_train)*100))
print("linear_svc_l1.score test: {:.2f}%\n".format(linear_svm_l1.score(X_test_scaled, y_test)*100))

np_resuts = np.around(np.array(linear_svm_l1.coef_), decimals=2)
data_pd = pd.DataFrame(data=np_resuts, index=['coef L1'])
#data_pd.columns = params
display(data_pd)

linear_svm_l2 = LinearSVC(C = 1, penalty="l2", dual=False)
linear_svm_l2.fit(X_train_scaled,y_train)
print("linear_svc_l2.score train: {:.2f}%".format(linear_svm_l2.score(X_train_scaled, y_train)*100))
print("linear_svc_l2.score test: {:.2f}%\n".format(linear_svm_l2.score(X_test_scaled, y_test)*100))

np_resuts = np.around(np.array(linear_svm_l2.coef_), decimals=2)
data_pd = pd.DataFrame(data=np_resuts, index=['coef L2'])
#data_pd.columns = params
display(data_pd)
```

```python
# MLP best
import mglearn
from sklearn.neural_network import MLPClassifier
import numpy as np
import matplotlib.pyplot as plt

max_score_test = 0
training_accuracy = []
test_accuracy = []
alpha2 = []
for activation in ['relu', 'identity', 'logistic','tanh']:
    for alpha_100 in range(1, 1000, 50):
        for max_iter in range(13, 15, 1):
            for hidden_layer_sizes in range(13, 15, 1):
                mlp = MLPClassifier(solver="lbfgs", random_state = 1, activation=activation, alpha=alpha_100/1000, max_iter=max_iter, hidden_layer_sizes=hidden_layer_sizes)
                #mlp = MLPClassifier(solver="lbfgs", random_state = 1, hidden_layer_sizes=hidden_layer_sizes)
                mlp.fit(X_train_scaled, y_train)
                train = mlp.score(X_train_scaled, y_train)*100
                test = mlp.score(X_test_scaled, y_test)*100
                training_accuracy.append(train)
                test_accuracy.append(test)
                alpha2.append(hidden_layer_sizes)
                if (max_score_test < test and train ==100) :
                #if (test > 96.59 and flag == False):
                    max_score_test = test
                    score_train = train
                    best_activation = activation
                    best_alpha = alpha_100/1000
                    best_max_iter = max_iter
                    best_hidden_layer_sizes = hidden_layer_sizes
print("score_train: ", score_train)
print("best_score_test: ", max_score_test)
print("best_activation: ", best_activation)
print("best_alpha: ", best_alpha)
print("best_max_iter", best_max_iter)
print("best_hidden_layer_sizes", hidden_layer_sizes)

#plt.plot(alpha2, training_accuracy, label = "Training")
#plt.plot(alpha2, test_accuracy, label = "Test")
#plt.ylabel("Accuracy")
#plt.xlabel("hidden_layer_sizes")
#plt.legend()

np_resuts = np.array(np.around(mlp.predict_proba(X_test_scaled), decimals=1))
data_pd = pd.DataFrame(np_resuts)
data_pd.columns = ['g','b']
display(data_pd)
```

```python
# GradientBoostingClassifier best
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import train_test_split
import mglearn
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

max_score_test = 0
training_accuracy = []
test_accuracy = []
for n_estimators in range(1, 100, 2):
    for learning_rate_100 in range(1, 100, 5):
        for subsample in range(1, 10, 1):
                boosting = GradientBoostingClassifier(random_state = 1, learning_rate=learning_rate_100/100, n_estimators=n_estimators, subsample=subsample/10)
                boosting.fit(X_train_scaled, y_train)
                train = boosting.score(X_train_scaled, y_train)*100
                test = boosting.score(X_test_scaled, y_test)*100
                training_accuracy.append(train)
                test_accuracy.append(test)
                if (max_score_test < test and train ==100) :
                    max_score_test = test
                    score_train = train
                    best_learning_rate = learning_rate_100/100
                    best_subsample = subsample/10
                    best_n_estimators = n_estimators
print("score_train: ", score_train)
print("best_score_test: ", max_score_test)
print("best_n_estimators: ", best_n_estimators)
print("best_learning_rate: ", best_learning_rate)
print("best_subsample", best_subsample)

#plt.plot(alpha2, training_accuracy, label = "Training")
#plt.plot(alpha2, test_accuracy, label = "Test")
#plt.ylabel("Accuracy")
#plt.xlabel("hidden_layer_sizes")
#plt.legend()

np_resuts = np.array(np.around(boosting.predict_proba(X_test_scaled), decimals=1))
data_pd = pd.DataFrame(np_resuts)
data_pd.columns = ['g','b']
display(data_pd)
```

```python
# MultinomialNB best
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
import mglearn
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

MMS = MinMaxScaler(copy=True, feature_range=(0,1))
MMS.fit(X_train)
X_train_scaled = MMS.transform(X_train)
X_test_scaled = MMS.transform(X_test)

max_score_test = 0
training_accuracy = []
test_accuracy = []
for fit_prior in [True,False]:
    for alpha_1000 in range(1, 1000, 50):
                multinomialnb = MultinomialNB(fit_prior=fit_prior, alpha=alpha_1000/1000)
                multinomialnb.fit(X_train_scaled, y_train)
                train = multinomialnb.score(X_train_scaled, y_train)*100
                test = multinomialnb.score(X_test_scaled, y_test)*100
                training_accuracy.append(train)
                test_accuracy.append(test)
                if (max_score_test < test and train ==100) :
                    max_score_test = test
                    score_train = train
                    best_alpha = alpha_1000/1000
                    best_fit_prior = fit_prior
print("score_train: ", score_train)
print("best_score_test: ", max_score_test)
print("best_fit_prior: ", best_fit_prior)
print("best_best_alpha: ", best_alpha)

np_resuts = np.array(np.around(multinomialnb.predict_proba(X_test_scaled), decimals=1))
data_pd = pd.DataFrame(np_resuts)
data_pd.columns = ['g','b']
display(data_pd)
```
